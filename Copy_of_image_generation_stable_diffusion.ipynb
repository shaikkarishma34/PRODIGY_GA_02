```python
import torch
from diffusers import StableDiffusionPipeline

model_id = "runwayml/stable-diffusion-v1-5"

pipe = StableDiffusionPipeline.from_pretrained(
    model_id,
    torch_dtype=torch.float16
)

pipe = pipe.to("cuda")
```

```python
prompt = "A futuristic AI robot learning machine learning, digital art"
image = pipe(prompt).images[0]
image
```

```python
prompts = [
    "A cyberpunk city at night with neon lights, ultra realistic",
    "A cute robot studying in a library, cartoon style",
    "A peaceful mountain landscape at sunrise, digital painting"
]

for p in prompts:
    image = pipe(p).images[0]
    display(image)
```
